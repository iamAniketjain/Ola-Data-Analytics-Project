```python
data = pd.read_excel(r"C:\Users\Administrator\Documents\DA_Project_1\Bookings-100000-Rows.xlsx")
```

```python
data.isnull().sum()
```

```python
data.V_TAT.fillna(data['V_TAT'].mean(), inplace=True)
data.C_TAT.fillna(data['C_TAT'].mean(), inplace=True)
data.Canceled_Rides_by_Customer.fillna(data['Canceled_Rides_by_Customer'].mode()[0], inplace=True)
data.Canceled_Rides_by_Driver.fillna(data['Canceled_Rides_by_Driver'].mode()[0], inplace=True)
data.Incomplete_Rides.fillna(data['Incomplete_Rides'].mode()[0], inplace=True)
data.Incomplete_Rides_Reason.fillna(data['Incomplete_Rides_Reason'].mode()[0], inplace=True)
data.Payment_Method.fillna(data['Payment_Method'].mode()[0], inplace=True)
data.Driver_Ratings.fillna(data['Driver_Ratings'].mean(), inplace=True)
data.Customer_Rating.fillna(data['Customer_Rating'].mean(), inplace=True)
```
